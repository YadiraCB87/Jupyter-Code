```python
## Ubicacion de los Alojamientos

import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

##Llamar dataset

paises=pd.read_excel(r"C:\Users\ViG\Desktop\Ubication.xlsx")
#paises.head()

## Crear diccionario de colores para cada zona registrada

colors = {'Serbia': 'Green', 'Monaco': 'Pink', 'Australia': 'Purple', 'Belgium': 'Blue', 'France': 'Orange', 'Germany': 'Green', 'Malaysia': 'Green', 'Netherlands': 'Black'}

##Mapear las ubicaciones  de cada alojamiento con el color respectivo
ubicacion_color=paises.host_location.map(colors)


# Utiliza el diccionario de colores junto con el argumento 'c'

##Creacion grafico
#plt.figure(figsize=(18,8))
#plt.scatter(paises.latitude,paises.longitude,color=ubicacion_color)
#plt.title("Ubicacion de los Alojamientos")
#plt.xlabel("Latitud")
#plt.ylabel("Longitud")
#plt.show()
```

```python
## Agrupacion de datos de acuerdo a su Ubicacion

cant_ubicacion = paises.groupby('host_location')['host_location'].count()

top_10_ubicaciones = cant_ubicacion.nlargest(4)

##Crear las variables que contengan los valores del series creado
##ubicaciones=cant_ubicacion.index
##cantidad=cant_ubicacion.values
ubicaciones = top_10_ubicaciones.index
cantidad = top_10_ubicaciones.values

#creacion grafico
plt.figure(figsize=(18,8))
plt.bar(ubicaciones,cantidad, color=colors.values())
plt.xticks(ubicaciones)
plt.title("Ubicacion de los Alojamientos por ubicacion",size=15)
plt.xlabel("host_location", size=14)
plt.ylabel("#Alojamientos",size=14)
plt.show()
```

```python
## Agrupacion datos segun su ubicacion

precio_alojamientos_ubicacion=paises.groupby('host_location')['price'].mean()

top_10_precios = precio_alojamientos_ubicacion.nlargest(5)
                                                                
##crear variables que contienen los valores del series creado
                                                                
localidades=top_10_precios.index
precios=top_10_precios.values
                                                                
##creacion grafico
plt.figure(figsize=(16,8))
plt.bar(localidades,precios, color=colors.values())
plt.xticks(localidades)
plt.title("Precio segun zona",size=20)
plt.xlabel("Ubicacion", size=18)
plt.ylabel("Precio",size=14)
plt.show()  

```

```python
## Graficos de barra siempre inician en 0

plt.figure(figsize=(30,6))

plt.subplot(132)
plt.scatter(localidades, precios)
plt.ylabel("Precio", size=15)
plt.title('Grafico de puntos precio',size=16)


plt.subplot(133)
plt.plot(localidades, precios,color='r')
plt.ylabel("Precio", size=15)
plt.title('Grafico de lineas precio',size=16)
plt.show()
```

```python
##Analisis variable superhost, calificacion de alojamiento superhost o no superhost 

import matplotlib.pyplot as plt
```

```python
##calcular la cantidad de superhost y no superhost

superh=paises[paises.host_listings_count==1]
no_superh=paises[paises.host_listings_count==0]

print("Cantidad de SuperHost: "+str(superh.shape[0]))
print("Cantidad de No SuperHost: "+str(no_superh.shape[0]))

```

```python
cantidades=[superh.shape[0],no_superh.shape[0]]
etiquetas = ['SuperHost', 'No SuperHost']

## creacion de grafico de pastel

colores = ['#FFFF00', '#0000FF']  # Amarillo y Azul en formato hexadecimal

plt.figure(figsize=(16,7))
plt.title("Alojamientos SuperHost")
plt.pie(cantidades,labels = etiquetas,autopct='%1.1f%%',colors=colores)
plt.show()
```

```python
## Para comparar la disponibilidad

media_disponibilidad=paises.availability_365.mean()

## Creacion de nueva columna

paises["Diferencia_Disponibilidad"]=paises["availability_365"]-media_disponibilidad
```

```python
## Crear una variable para almacenar la media disponibile para cada zona

dif_disponibilidad=paises.groupby("host_location")[["Diferencia_Disponibilidad"]].mean()

##Filtrar las zonas con mayor disponibilidad comparando con el promedio
positivos=dif_disponibilidad[dif_disponibilidad.Diferencia_Disponibilidad<0]

##Filtrar las zonas con menos disponibilidad comparado con el promedio 
negativos=dif_disponibilidad[dif_disponibilidad.Diferencia_Disponibilidad<0]

##Creacion del grafico
plt.figure(figsize=(16,6))
plt.bar(positivos.index,positivos.Diferencia_Disponibilidad,color='#800000')
plt.bar(negativos.index,negativos.Diferencia_Disponibilidad,color='r')
plt.show()
```

```python
##Calificaciones , buscar el promedio de los rating de comunicacion, limpieza y check in

#Agrupar los resultados segun la ubicacion y se calcula la media de las calificaciones

mejores_calificados=paises.groupby("host_location")[["review_scores_rating", "review_scores_communication","review_scores_cleanliness","review_scores_checkin"]].mean()
                                                    
## Ordenar de manera descendete segun el valor del review_scores_rating
                                                    
mejores_calificados.sort_values(by=["review_scores_rating"],ascending=False)
```
