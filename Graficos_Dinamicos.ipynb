```python
##Crear Path
path= r"C:\Users\ViG\Desktop\DataSets"

units_sales = pd.read_csv(path+'\Dinamica.csv')#, sep=';', header=0,index_col=False,encoding='latin-1',
                         #engine='python')

units_sales.head()
```

```python
units_sales.set_index('Order Date', inplace = True) ## Definir Order Date como Indice
```

```python
## Crear nueva variable ## se acumula venta por region acum es acumular y sumar

units_sales_acum = units_sales.cumsum(axis=0)
```

```python
units_sales_acum 
```

```python
## Grafico Dinamico 

bcr.bar_chart_race(df = units_sales_acum, filename = None, 
                  figsize =(3.5, 3), title = 'Venta Acumulada Ao 2014')
```
