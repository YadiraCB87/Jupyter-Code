```python
graficos1_1 = pd.crosstab(copia2.Marital_Status,copia2.Column1)
graficos1_1.plot.barh()
```

```python
graficos1 = pd.crosstab(copia2.Column1,copia2.Education)
graficos1.plot.barh()
```

```python
datos_porcentuales_grafico = graficos1.apply(lambda x: (x*100) /sum(x),axis=1)
datos_porcentuales_grafico.plot.barh(stacked=True)
```
